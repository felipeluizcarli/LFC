```python
import matplotlib.pyplot as plt

# Dados de exemplo
x = [1, 2, 3, 4, 5]
y = [1.5, 4, 6.4, 8, 10]

# Criar o gráfico de linha
plt.plot(x, y)

# Personalizar o gráfico
plt.title('Gráfico de Linha')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

# Mostrar o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dados de exemplo
categorias = ['A', 'B', 'C', 'D']
valores = [10, 20, 15, 25]

# Criar o gráfico de barras
plt.bar(categorias, valores)

# Personalizar o gráfico
plt.title('Gráfico de Barras')
plt.xlabel('Categorias')
plt.ylabel('Valores')

# Mostrar o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dados de exemplo
labels = ['A', 'B', 'C', 'D']
valores = [30, 15, 20, 35]

# Criar o gráfico de pizza
plt.pie(valores, labels=labels)

# Personalizar o gráfico
plt.title('Gráfico de Pizza')

# Mostrar o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dados de exemplo
valores = [1, 2, 2, 3, 3, 3, 4, 4, 5]

# Criar o histograma
plt.hist(valores)

# Personalizar o gráfico
plt.title('Histograma')
plt.xlabel('Valores')
plt.ylabel('Frequência')

# Mostrar o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados de exemplo
categorias = ['A', 'B', 'C', 'D']
valores = [0.6, 0.8, 0.2, 0.4]

# Criar o gráfico de radar
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
theta = np.linspace(0, 2 * np.pi, len(categorias), endpoint=False).tolist()
theta += theta[:1]
valores += valores[:1]
ax.plot(theta, valores)

# Personalizar o gráfico
ax.fill(theta, valores, alpha=0.25)
ax.set_xticks(theta[:-1])
ax.set_xticklabels(categorias)
ax.set_title('Gráfico de Radar')

# Mostrar o gráfico
plt.show()
```
