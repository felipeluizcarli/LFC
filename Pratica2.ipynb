```python
meses = {1: 'Janeiro', 2:'Fevereiro', 3: 'Março'}
```

Dicionario python

```python
df['Data'] = pd.to_datetime(df['Data']).dt.month.map(meses)
```

Este é um trecho de código Python que converte uma coluna de datas em um DataFrame do Pandas para o nome do mês correspondente, utilizando um dicionário meses previamente definido.

```python
df.head()
```

Exibr tabela

```python
df['Horário'] = df['Horário'].dt.strftime('%H')
```

Este é um trecho de código Python que converte uma coluna de horários em um formato datetime em um DataFrame do Pandas para uma string contendo apenas as horas do horário.

```python
df.head()
```

```python
time = [] #lista vazia
for i in list(map(int,df['Horário'])):
    if i > 8 and i < 13:
      time.append('Manhã') 
    elif i > 12 and i < 19: 
      time.append('Tarde')
    else:
      time.append('Noite')
df['Horário'] = time
```
